### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row)
```

```python
# 1. 取出班次一的每一個時間
with open('./Data/example.csv', newline='', encoding='utf-8') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        print(row[5])
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存
trainOne = {}
with open('./Data/example.csv', newline='', encoding='utf-8') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        if not row[0].isdecimal(): continue
        trainOne.update({row[0]: row[5]})
    print(trainOne)
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

trainsNorm = {}
trainsWeek = {}
with open('./Data/example.csv', newline='', encoding='utf-8') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        if not row[0].isdecimal(): continue
        for index, value in enumerate(row):
            if index > 4 and index < 16:
                if not index - 4 in trainsNorm:
                    trainsNorm.update({index - 4: {}})
                    trainsWeek.update({index - 4: {}})
                if row[3] == '平日':
                    trainsNorm[index - 4].update({row[0]: value})
                elif row[3] == '假日':
                    trainsWeek[index - 4].update({row[0]: value})
for i in range(1, 12):
    print('平日 班次{}的時間:{}'.format(i, trainsNorm[i]))
    print('假日 班次{}的時間:{}'.format(i, trainsWeek[i]))
```
