## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.dcard.tw/f'
resp = requests.get(url)
soup = BeautifulSoup(resp.text)
soup
```

## Grab + PyQuery

```python
from grab import Grab
gb = Grab()
resp = gb.go(url)
resp.body
```

```python
from pyquery import PyQuery as pq
doc = pq(url)
doc.html()
```
