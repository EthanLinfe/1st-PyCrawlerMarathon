## 2. 取出現在時間兩小時內的新聞

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
from datetime import datetime
from datetime import timedelta
url = 'https://www.ettoday.net/news/news-list.htm'
now = datetime.now()
hour = now + timedelta(hours=-2)
done = False

chrome_option = webdriver.chrome.options.Options()
chrome_option.add_argument('--headless')
browser = webdriver.Chrome(executable_path='chromedriver',chrome_options=chrome_option)
browser.get(url)
while True:
    time.sleep(0.5)
    browser.execute_script('window.scrollBy(0, 3000);')
    r = browser.page_source
    soup = BeautifulSoup(r, 'html.parser')
    news = soup.find_all(class_='part_list_2')
    for new in news[0].find_all('h3'):
        t1 = new.find('span').text.split(' ')
        t2 = t1[0].split('/')
        t3 = t1[1].split(':')
        curr = datetime(year=int(t2[0]), month=int(t2[1]), day=int(t2[2]), hour=int(t3[0]), minute=int(t3[1]))
        if curr < hour:
            done = True
            break
        print(new.text)
    if done: break
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
from datetime import datetime
from datetime import timedelta
url = 'https://www.ettoday.net/news/news-list.htm'
now = datetime.now()
day = now + timedelta(days=-3)
day1 = day.replace(hour=15, minute=0,second=0,microsecond=0)
day2 = day.replace(hour=17, minute=0,second=0,microsecond=0)
start = False
done = False
chrome_option = webdriver.chrome.options.Options()
chrome_option.add_argument('--headless')
browser = webdriver.Chrome(executable_path='chromedriver',chrome_options=chrome_option)
browser.get(url)
while True:
    time.sleep(0.5)
    browser.execute_script('window.scrollBy(0, 10000);')
    r = browser.page_source
    soup = BeautifulSoup(r, 'html.parser')
    news = soup.find_all(class_='part_list_2')
    for new in news[0].find_all('h3'):
        t1 = new.find('span').text.split(' ')
        t2 = t1[0].split('/')
        t3 = t1[1].split(':')
        curr = datetime(year=int(t2[0]), month=int(t2[1]), day=int(t2[2]), hour=int(t3[0]), minute=int(t3[1]))
        print(curr)
        if curr < day2:
            start = True
        if curr > day1 and start:
            done = True
        if start: print(new.text)
    if done: break
browser.quit()
```
